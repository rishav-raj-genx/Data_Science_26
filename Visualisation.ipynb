```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = [1,2,3,4,5,6]
y = [1,4,9,8.6,4.5,1]

plt.figure(figsize=(10,6))
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('V__D')
# plt.show()
```

```python
subjects = ['Maths', 'Physics', 'CSE', 'EE']
scores = [88, 76, 95, 29]
plt.bar(subjects, scores, color = 'blue', edgecolor = 'black')
plt.title('Student Scores')
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.plot('Maths', 88, 'ro')
plt.plot('Physics', 76, 'ro')
plt.plot('CSE', 95, 'ro')
plt.plot('EE', 29, 'ro')
```

```python
hours = np.random.randint(1,10,50)
marks = hours * 8 + np.random.randint(1,20,50)
plt.scatter(hours, marks, s=10)
```
