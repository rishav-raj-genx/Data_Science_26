```python
import numpy as np
n = input("Enter cricket scores separated by spaces (e.g., 45 102 12 88): ")
cri = [int(num) for num in n.split()]
cric = np.array(cri)
print("\n--- Tournament Scorecard ---")
print(" Day   |  Score ")
print("----------------")
for index, score in enumerate(cric):
    print(f" Day {index + 1} |   {score}")
print("----------------")
print("\n--- Player Statistics ---")
print(f"Original Array:     {cric}")
print(f"Highest Score:      {np.max(cric)}")
print(f"Lowest Score:       {np.min(cric)}")
print(f"Average (Mean):     {np.mean(cric):.2f}")
print(f"Standard Deviation: {np.std(cric):.2f}")
print(f"Unique Scores:      {np.unique(cric)}")
print(f"Sorted Scores:      {np.sort(cric)}")
```

A shop - we need a price input from the user and find the max price, stand deviation, mini price, total price, mean

```python
import numpy as np
items = []
prices_of_the_respective_item = []
for i in range(7):
    name = input(f"Enter name of item {i+1}: ")
    price = float(input(f"Enter price of {name}: "))
    items.append(name)
    prices_of_the_respective_item.append(price)
prices = np.array(prices_of_the_respective_item)
print(prices)
max_price_product = np.max(prices)
min_price_product = np.min(prices)
mean_price_of_products = np.mean(prices)
std_price_of_products = np.std(prices)
max_item = items[np.argmax(prices)]
min_item = items[np.argmin(prices)]
print("\n----- Results -----")
print(f"Most expensive item : {max_item} (₹{max_price_product})")
print(f"Cheapest item       : {min_item} (₹{min_price_product})")
print(f"Average price       : ₹{mean_price_of_products:.2f}")
print(f"Standard Deviation  : ₹{std_price_of_products:.2f}")
```
