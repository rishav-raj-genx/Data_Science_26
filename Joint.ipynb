```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    'id': [1,2,3,4],
    'name': ['Raja', 'Rahul', 'Rhea', 'Rani'],
    'age': [23, 24, 27, 29]
})

df2 = pd.DataFrame({
    'id': [3,4,5,6],
    'name': ['Rhea', 'Rani', 'Rihana', 'Rupesh'],
    'age': [27, 29, 26, 20]
})
```

```python
pd.merge(df1, df2, on = 'id', how = 'inner')
```
