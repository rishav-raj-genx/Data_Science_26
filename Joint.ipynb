```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    'id': [1,2,3,4],
    'name': ['Raja', 'Rahul', 'Rhea', 'Rani'],
    'age': [23, 24, 27, 29]
})

df2 = pd.DataFrame({
    'id': [3,4,5,6],
    'name': ['Rhea', 'Rani', 'Rihana', 'Rupesh'],
    'age': [27, 29, 26, 20]
})
```

```python
df3 = pd.merge(df1, df2, on = 'id', how = 'inner')
```

```python
df3.rename(columns={'name_x': 'table_name1'}, inplace=True)
print(df3)
```

```python
df4 = pd.merge(df1, df2, on = 'id', how = 'left')
print(df4)
```

```python
df5 = pd.merge(df1, df2, on = 'id', how = 'right')
print(df5)
```

```python
df6 = pd.merge(df1, df2, on = 'id', how = 'outer')
print(df6)
```

```python
df7 = pd.merge(df1, df2, how = 'cross')
display(df7)
```
